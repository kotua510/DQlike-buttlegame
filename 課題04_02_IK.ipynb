# **ドラクエっぽいターン制バトルゲーム**

## **概要**

このプログラムは、ドラゴンクエストを意識して作った交互に敵とプレイヤーが行動を行うことによって敵を倒すターン制バトルゲームのプログラムです。  プレイヤーの行動はプレイヤー自身が決め、敵の行動はプログラムによりランダムに決定されます。    

## **操作方法**

まず、このプログラムは二つのコードファイルからできていますが、**まず上の「りゅうキング(敵)第一形態 戦闘」のコードファイルから実行してください**、このプログラムは「りゅうキング(敵)第一形態 戦闘」で必要な変数を定義しているので先にこちらを実行しないと変数が定義でぎず、「りゅうキング(敵)第二形態 戦闘」にてエラーになってしまいます。

また、コマンドやプレイヤーネームは入力を求めますが**名前、使いたい道具は全角で言葉を,コマンド、使いたい呪文、難易度は半角の番号で入力してください**。

入力する際、**入力してから次の入力が速すぎると、プログラムに入力されてしまうので注意してください**

このプログラムは2つのコードファイルからできていますので、2つ目を忘れないで実行してください。

HPは体力を表し、MPはマジックポイントを表します。  自分のHPが0になればゲームオーバーですし、敵のHPを0にすればゲームクリアです。MPは呪文を使用につかうポイントです、敵も自分も呪文の使用時にMPを消費し呪文に設定されているMPを自分や敵のMPが下回っている時は呪文はつかえません
